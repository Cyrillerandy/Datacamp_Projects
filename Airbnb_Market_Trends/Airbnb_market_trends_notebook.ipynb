**Load the data first.**

```python
# load the datasets
# csv
airbnb_price = pd.read_csv('data/airbnb_price.csv', index_col = 0)

# tsv
airbnb_last_review = pd.read_csv('data/airbnb_last_review.tsv', sep = '\t', index_col = 0)
```

```python
# view reviews 
airbnb_last_review.head()
```

```python
# load the excel file
file = pd.ExcelFile('data/airbnb_room_type.xlsx')
print(file.sheet_names)

# load dataset
airbnb_room_type = file.parse(0).set_index('listing_id')
airbnb_room_type.head()
```

**Q1. What are the dates of the earliest and most recent reviews? Store these values as two separate variables with your preferred names.**

```python
# view reviews
airbnb_last_review.head()
```

```python
# convert last review col to datetime format and order 
last_review = airbnb_last_review.copy()
last_review['date'] = pd.to_datetime(last_review['last_review'], errors = 'coerce', infer_datetime_format = True)
last_review.head()
```

```python
# find the earliest and most recent dates
earliest_review = last_review['date'].min().strftime('%Y-%m-%d')
latest_review = last_review['date'].max().strftime('%Y-%m-%d')

print(f'The earliest review date was {earliest_review} and the most recent review date was {latest_review}')
```

**Q2. How many of the listings are private rooms? Save this into any variable.**

```python
# view airbnb room types
airbnb_room_type.head()
```

```python
# make a copy and view the unique room types
room_type = airbnb_room_type.copy()

room_type['room_type'].unique()
```

We clearly see the room types have formatting issues with capitalisation, fix it then recheck

```python
# Fix the errors
room_type['room_type'] = room_type['room_type'].str.lower()

# review the changes
room_type['room_type'].unique()
```

```python
# get the number of private listings
room_type['room_type'].value_counts().values
```

```python
# save it to a variable
private_rooms = room_type['room_type'].value_counts().values[1]

print(f'The number of listings that are private rooms are {private_rooms}')
```

**Q3. What is the average listing price? Round to the nearest two decimal places and save into a variable.**

```python
# view prices data
airbnb_price.head()
```

```python
# make copy 
price = airbnb_price.copy()

# check dtypes
price.dtypes
```

```python
# clean the dtype for price
price['price'] = price['price'].replace(to_replace = r' dollars', value = '', regex = True).str.strip()
price['price'] = price['price'].astype('int')

price.dtypes
```

```python
# find the average
average_listing_price = round(price['price'].mean(), 2)

print(f'The average listing price is $ {average_listing_price}')
```

**Q5. Combine the new variables into one DataFrame called review_dates with four columns in the following order: first_reviewed, last_reviewed, nb_private_rooms, and avg_price. The DataFrame should only contain one row of values.**

```python
# make the dict to hold the df values
review_dict = {'first_reviewed' : [pd.to_datetime(earliest_review)], 'last_reviewed' : [pd.to_datetime(latest_review)],
              'nb_private_rooms' : [private_rooms], 'avg_price' : [average_listing_price]}

# make a df
review_dates = pd.DataFrame(review_dict)

# view df
review_dates
```
