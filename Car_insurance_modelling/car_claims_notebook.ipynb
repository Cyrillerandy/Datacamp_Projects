```python
# Import required modules
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit

# Start coding!
```

### Identify the single feature of the data that is the best predictor of whether a customer will put in a claim (the "outcome" column), excluding the "id" column.

```python
# load the data
cars = pd.read_csv('car_insurance1.csv')

cars.head()
```

```python
# Get the explanatory variable column names
exp_cols = [col for col in cars.columns if col not in ['id', 'outcome']]

print(exp_cols)
```

```python
# reponse variable column
resp = 'outcome'
```

```python
# fit the models on each pair of exp vs resp
# first create a dict to hold confusion matrices
conf_mats = {}

# loop through the explanatory variables fitting each 
# and store the matrix
for exp in exp_cols:
    conf_mats[exp] = logit(formula = resp + ' ~ ' + exp, data = cars).fit().pred_table()
    
# Check if the assignment worked correctly
print('The matrices :\n', conf_mats)
```

```python
# Get the accuracies 
# Create empty dict
accuracies = {}

# loop through and calculate
for k, v in conf_mats.items():
    accuracies[k] = [(v[0, 0] + v[1, 1]) / (v[0, 0] + v[1, 0] + v[0, 1] + v[1, 1])]
    
print('Accuracies :\n', accuracies)
```

```python
# make a dataframe of the accuracies
accuracy_df = pd.DataFrame(data = accuracies).T.reset_index()
accuracy_df = accuracy_df.rename(columns = {'index' : 'best_feature', 0:'best_accuracy'})
accuracy_df.sort_values(by = 'best_accuracy', ascending = False, inplace = True)
accuracy_df.reset_index(drop = True, inplace = True)
accuracy_df
```
