### QUESTION 1

What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called `duration`.

```python
# filter the data for release year between 1990 and 1999 (the 90s)
import numpy as np

df_90s = netflix_df[np.logical_and(netflix_df['release_year'] >= 1990, 
                                  netflix_df['release_year'] <= 1999)]
# preview the data
print(df_90s.shape)
df_90s
```

```python
# Find the most frequent movie duration
duration = df_90s['duration'].value_counts().index[0]
print(f'The most frequent movie duration was {duration} minutes')
```

### QUESTION 2 

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# first filter the df on genre to only have action movies
action_90s = df_90s[df_90s['genre'] == 'Action']
# view data
print(action_90s.shape)
action_90s
```

```python
# Now filter for short action movies (less than 90 minutes)
short_action_movies = action_90s[action_90s['duration'] < 90]
# view the data
print(short_action_movies.shape)
short_action_movies
```

```python
# Save the number
short_movie_count = len(short_action_movies)
print(f'The number of short action movies in the 90s is {short_movie_count}')
```
